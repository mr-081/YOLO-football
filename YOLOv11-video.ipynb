```python
import cv2
```

```python
from ultralytics import YOLO
```

```python
import matplotlib.pyplot as plt
```

```python
# Load a COCO-pretrained YOLO11n model
model = YOLO("yolo11n.pt")
```

```python
# Open the video file
video_path = 'Football.mp4'  # Path to your video file
cap = cv2.VideoCapture(video_path)
```

```python
width = 480
height = 320
fourcc = cv2.VideoWriter_fourcc(*'mp4v')    # Codec for mp4
out = cv2.VideoWriter('Football_O.mp4', fourcc, 10.0, (width, height))
```

```python
# Check if the video file opened successfully
if not cap.isOpened():
    print("Error: Could not open video.")
    exit()
```

```python
# Loop over the frames from the video
while True:
    ret, frame = cap.read()

    # If the frame was not retrieved successfully (end of video), break the loop
    if not ret:
        break

    # Resize the frame if needed (optional)
    resized_frame = cv2.resize(frame, (480, 320))

    # Run inference on the resized frame
    results = model(resized_frame)

    # Plot the frame with bounding boxes
    result_frame = results[0].plot()

    # Convert the frame to RGB (OpenCV uses BGR)
    result_frame_rgb = cv2.cvtColor(result_frame, cv2.COLOR_BGR2RGB)

    #### Write the frame with detections to the output video
    out.write(result_frame)

    # Display the result frame using cv2_imshow for Colab
    cv2.imshow('frame', result_frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture object and close any OpenCV windows
cap.release()
out.release()
cv2.destroyAllWindows()
```
